```python
import pandas as pd
import os
import numpy as np

import arcpy

#import arcgis libraries
from arcgis.gis import *
gis = GIS()
```

```python
# IMPORTANT!!! download and save to data folder : https://gis.utah.gov/data/water/wetlands/
dirWLData = r"data\UT_shapefile_wetlands"
fnWLshp = 'UT_Wetlands.shp'
fnRPshp = 'UT_Riparian.shp'
dWLFilterField = 'WETLAND_TY'
dWLFilterValues_Out = 'Riverine'
```

```python
sdfWL = pd.DataFrame.spatial.from_featureclass(os.path.join(dirWLData,fnWLshp))
sdfWL = sdfWL[sdfWL[dWLFilterField]!=dWLFilterValues_Out]
sdfWL
```

```python
sdfRP = pd.DataFrame.spatial.from_featureclass(os.path.join(dirWLData,fnRPshp))
sdfRP
```

```python
sdfWL.spatial.to_featureclass(os.path.join('intermediate','wetlands.shp' ))
sdfRP.spatial.to_featureclass(os.path.join('intermediate','riparians.shp'))
```

```python
output_wl_fc = os.path.join('intermediate','simplified_wetlands.shp')
simplify_tolerance = "100 Feet"

# use the Simplify Polygon tool to simplify the geometry of the input feature class
arcpy.cartography.SimplifyPolygon(os.path.join('intermediate','wetlands.shp' ), output_wl_fc, "POINT_REMOVE", simplify_tolerance)
```
